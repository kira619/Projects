# Logistic Regression

Now it's time to do a train test split, and train our model!

You'll have the freedom here to choose columns that you want to train on!

** Split the data into training set and testing set using train_test_split**

```python
from sklearn.model_selection import train_test_split
df=df.drop(['Ad Topic Line', 'City','Country','Timestamp'],axis=1)
```

```python
x=df.drop('Clicked on Ad',axis=1)
y=df['Clicked on Ad']
sns.heatmap(data=df.isnull(),yticklabels=False,cmap='rainbow')
#Hence,No missing Data
```

```python
x
```

** Train and fit a logistic regression model on the training set.**

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=101)
ytest
```

```python
from sklearn.linear_model import LogisticRegression
logmodel= LogisticRegression(solver='lbfgs',class_weight='balanced', max_iter=100000)
logmodel.fit(xtrain,ytrain)

```

## Predictions and Evaluations
** Now predict values for the testing data.**

```python
predictions=logmodel.predict(xtest)
predictions
```

** Create a classification report for the model.**

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(ytest,predictions))
```

```python
print(confusion_matrix(ytest,predictions))
```
